```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read Image

```python
img = cv2.imread('nad.jpg')
plt.imshow(img)
```

### Gambar Asli

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

### Cropping Image

```python
crop_img = img[400:700,250:500]
plt.imshow(crop_img)
```

```python
kernel = [[0, 1, 0],
          [1, -4, 1],
          [0, 1, 0]]

blur = (1/9.0) * np.array([[1,1,1],
                           [1,1,1],
                           [1,1,1]])

sharpen = [[0, -1, 0],
          [-1,5,-1],
          [0,-1,0]]

kernel = np.array(sharpen)
```

### Gunakan Filter Sharpen (Sharpen Filter)

```python
hasil = cv2.filter2D(crop_img, -1, kernel)

fig, ax = plt.subplots(1,2, figsize=(25,8))

ax[0].imshow(crop_img)
ax[0].set_title('Tanpa Filter')
ax[1].imshow(hasil)
ax[1].set_title('Filter')

plt.imshow(hasil)
```

### Grayscale to Black White (biner)

```python
gray = cv2.cvtColor(crop_img, cv2.COLOR_RGB2GRAY)
(thresh, bw) = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
plt.imshow(bw, cmap='gray')
```

### Save Image

```python
cv2.imwrite('hasil_nad.jpg', bw)
```
